# 2. Ventas por categor√≠a

En este debes calcular la cantidad de productos vendidos por categor√≠a en cada tienda. La idea es agrupar los datos por categor√≠a y contar el n√∫mero de ventas de cada tipo, mostrando las categor√≠as m√°s populares de cada tienda.

En donde este debes calcular la cantidad de productos vendidos por categor√≠a en cada tienda. La idea es agrupar los datos por categor√≠a y contar el n√∫mero de ventas de cada tipo, mostrando las categor√≠as m√°s populares de cada tienda.

```python
import pandas as pd

## 1. Funci√≥n para contar productos por categor√≠a en una tienda
def productos_por_categoria(df_tienda, nombre_tienda):
    """
    Calcula la cantidad de productos vendidos por categor√≠a en una tienda

    Args:
        df_tienda (DataFrame): DataFrame de la tienda
        nombre_tienda (str): Nombre de la tienda

    Returns:
        DataFrame: Conteo de productos por categor√≠a
    """
    conteo = df_tienda['Categor√≠a del Producto'].value_counts().reset_index()
    conteo.columns = ['Categor√≠a', 'Cantidad Vendida']
    conteo['Tienda'] = nombre_tienda
    return conteo
```

```python
## 2. Procesamiento para todas las tiendas
# Lista para almacenar los resultados
resultados_categorias = []

# Procesamos cada tienda
resultados_categorias.append(productos_por_categoria(tienda,'Tienda 1'))
resultados_categorias.append(productos_por_categoria(tienda2, 'Tienda 2'))
resultados_categorias.append(productos_por_categoria(tienda3, 'Tienda 3'))
resultados_categorias.append(productos_por_categoria(tienda4, 'Tienda 4'))

# Combinamos todos los resultados
categorias_tiendas = pd.concat(resultados_categorias, ignore_index=True)
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Configuraci√≥n de estilo CORREGIDA (usando seaborn directamente)
sns.set_theme(style="whitegrid")  # Estilo moderno de seaborn
plt.rcParams['figure.facecolor'] = 'white'  # Fondo blanco para los gr√°ficos

# 1. Funci√≥n para analizar categor√≠as (mejorada)
def analizar_categorias(df, nombre_tienda, top_n=5):
    categorias = (
        df['Categor√≠a del Producto']
        .value_counts()
        .nlargest(top_n)
        .reset_index()
    )
    categorias.columns = ['Categor√≠a', 'Cantidad']
    categorias['Tienda'] = nombre_tienda
    categorias['Porcentaje'] = (categorias['Cantidad'] / categorias['Cantidad'].sum()) * 100
    return categorias

# 2. Procesamiento para todas las tiendas (con manejo de errores)
try:
    top_categorias = pd.concat([
        analizar_categorias(tienda, 'Tienda 1'),
        analizar_categorias(tienda2, 'Tienda 2'),
        analizar_categorias(tienda3, 'Tienda 3'),
        analizar_categorias(tienda4, 'Tienda 4')
    ])

    # 3. Visualizaci√≥n mejorada
    g = sns.FacetGrid(top_categorias, col='Tienda', col_wrap=2, height=5, aspect=1.5)
    g.map_dataframe(
        sns.barplot,
        x='Cantidad',
        y='Categor√≠a',
        palette='Blues_d'
    )

    # A√±adir t√≠tulos y etiquetas
    g.set_titles("{col_name}")
    g.set_axis_labels("Unidades Vendidas", "")

    # A√±adir valores en las barras
    for ax in g.axes.flat:
        for p in ax.patches:
            ax.annotate(
                f"{int(p.get_width()):,}\n({p.get_width()/top_categorias[top_categorias['Tienda']==ax.title.get_text()]['Cantidad'].sum():.1f}%)",
                (p.get_width(), p.get_y() + p.get_height()/2),
                ha='left', va='center', xytext=(5, 0), textcoords='offset points'
            )

    plt.suptitle('Top Categor√≠as por Tienda', y=1.02, fontweight='bold')
    plt.tight_layout()
    plt.show()

    # 4. Tabla resumen interactiva
    print("\nüìä Tabla Resumen de Categor√≠as Principales")
    display(
        top_categorias.pivot_table(
            index='Categor√≠a',
            columns='Tienda',
            values='Cantidad',
            aggfunc='sum',
            fill_value=0
        )
        .style
        .background_gradient(cmap='Blues', axis=1)
        .format('{:,.0f}')
        .set_caption('Unidades Vendidas por Categor√≠a y Tienda')
    )

except Exception as e:
    print(f"‚ùå Error en el an√°lisis: {e}")
    print("Verifica que los DataFrames (tienda, tienda2, etc.) est√©n correctamente cargados")
```
