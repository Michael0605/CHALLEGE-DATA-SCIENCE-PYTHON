En este debes calcular la cantidad de productos vendidos por categor√≠a en cada tienda. La idea es agrupar los datos por categor√≠a y contar el n√∫mero de ventas de cada tipo, mostrando las categor√≠as m√°s populares de cada tienda.

En donde este debes calcular la cantidad de productos vendidos por categor√≠a en cada tienda. La idea es agrupar los datos por categor√≠a y contar el n√∫mero de ventas de cada tipo, mostrando las categor√≠as m√°s populares de cada tienda.

```python
import pandas as pd

## 1. Funci√≥n para contar productos por categor√≠a en una tienda
def productos_por_categoria(df_tienda, nombre_tienda):
    """
    Calcula la cantidad de productos vendidos por categor√≠a en una tienda

    Args:
        df_tienda (DataFrame): DataFrame de la tienda
        nombre_tienda (str): Nombre de la tienda

    Returns:
        DataFrame: Conteo de productos por categor√≠a
    """
    conteo = df_tienda['Categor√≠a del Producto'].value_counts().reset_index()
    conteo.columns = ['Categor√≠a', 'Cantidad Vendida']
    conteo['Tienda'] = nombre_tienda
    return conteo
```

```python
## 2. Procesamiento para todas las tiendas
# Lista para almacenar los resultados
resultados_categorias = []

# Procesamos cada tienda
resultados_categorias.append(productos_por_categoria(tienda,'Tienda 1'))
resultados_categorias.append(productos_por_categoria(tienda2, 'Tienda 2'))
resultados_categorias.append(productos_por_categoria(tienda3, 'Tienda 3'))
resultados_categorias.append(productos_por_categoria(tienda4, 'Tienda 4'))

# Combinamos todos los resultados
categorias_tiendas = pd.concat(resultados_categorias, ignore_index=True)
```

# 3. Calificaci√≥n promedio de la tienda


```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Configuraci√≥n de estilos
sns.set_theme(style="whitegrid")
plt.figure(figsize=(10, 6))

## 1. C√°lculo de calificaciones promedio por tienda
def calcular_satisfaccion():
    # Diccionario para almacenar resultados
    resultados = {
        'Tienda': ['Tienda 1', 'Tienda 2', 'Tienda 3', 'Tienda 4'],
        'Calificaci√≥n Promedio': [
            tienda['Calificaci√≥n'].mean(),
            tienda2['Calificaci√≥n'].mean(),
            tienda3['Calificaci√≥n'].mean(),
            tienda4['Calificaci√≥n'].mean()
        ],
        'Mediana de Calificaci√≥n': [
            tienda['Calificaci√≥n'].median(),
            tienda2['Calificaci√≥n'].median(),
            tienda3['Calificaci√≥n'].median(),
            tienda4['Calificaci√≥n'].median()
        ],
        'Total Rese√±as': [
            len(tienda),
            len(tienda2),
            len(tienda3),
            len(tienda4)
        ]
    }

    return pd.DataFrame(resultados)

# Calculamos las m√©tricas
satisfaccion_df = calcular_satisfaccion().round(2)

## 2. Visualizaci√≥n de resultados
# Gr√°fico de barras para calificaciones promedio
plt.figure(figsize=(12, 6))
ax = sns.barplot(
    data=satisfaccion_df,
    x='Tienda',
    y='Calificaci√≥n Promedio',
    palette='viridis',
    errorbar=None
)

# A√±adir etiquetas de valor
for p in ax.patches:
    ax.annotate(
        f"{p.get_height():.2f}",
        (p.get_x() + p.get_width() / 2., p.get_height()),
        ha='center', va='center',
        xytext=(0, 10),
        textcoords='offset points'
    )

plt.title('Calificaci√≥n Promedio por Tienda', fontweight='bold', pad=20)
plt.xlabel('Tienda', labelpad=10)
plt.ylabel('Puntuaci√≥n Promedio (1-5)', labelpad=10)
plt.ylim(0, 5.5)  # La escala de calificaci√≥n es de 1 a 5
plt.tight_layout()
plt.show()

## 3. An√°lisis detallado
print("\nüîç An√°lisis Detallado de Satisfacci√≥n:")
display(satisfaccion_df.style
       .background_gradient(subset=['Calificaci√≥n Promedio'], cmap='RdYlGn', vmin=1, vmax=5)
       .format({'Calificaci√≥n Promedio': '{:.2f}', 'Mediana de Calificaci√≥n': '{:.2f}'})
       .set_caption('M√©tricas de Satisfacci√≥n del Cliente'))

## 4. Distribuci√≥n de calificaciones (boxplot)
plt.figure(figsize=(12, 6))
sns.boxplot(
    data=pd.concat([
        tienda.assign(Tienda='Tienda 1'),
        tienda2.assign(Tienda='Tienda 2'),
        tienda3.assign(Tienda='Tienda 3'),
        tienda4.assign(Tienda='Tienda 4')
    ]),
    x='Tienda',
    y='Calificaci√≥n',
    palette='Set2'
)
plt.title('Distribuci√≥n de Calificaciones por Tienda', fontweight='bold', pad=20)
plt.xlabel('Tienda', labelpad=10)
plt.ylabel('Calificaci√≥n', labelpad=10)
plt.ylim(0, 5.5)
plt.tight_layout()
plt.show()
```

# 4. Productos m√°s y menos vendidos

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from matplotlib.ticker import FuncFormatter

# Configuraci√≥n de estilos
sns.set_theme(style="whitegrid")
plt.rcParams['figure.facecolor'] = 'white'

# Funci√≥n para formatear n√∫meros grandes
def format_thousands(x, pos):
    return f'{int(x/1000)}K' if x >= 1000 else f'{int(x)}'

thousands_formatter = FuncFormatter(format_thousands)

## 1. Funci√≥n para analizar productos por tienda
def analizar_productos(df, nombre_tienda, top_n=5):
    # Productos m√°s vendidos
    top_productos = df['Producto'].value_counts().nlargest(top_n).reset_index()
    top_productos.columns = ['Producto', 'Ventas']
    top_productos['Tipo'] = 'M√°s vendidos'
    top_productos['Tienda'] = nombre_tienda

    # Productos menos vendidos
    bottom_productos = df['Producto'].value_counts().nsmallest(top_n).reset_index()
    bottom_productos.columns = ['Producto', 'Ventas']
    bottom_productos['Tipo'] = 'Menos vendidos'
    bottom_productos['Tienda'] = nombre_tienda

    return pd.concat([top_productos, bottom_productos])

## 2. Procesamiento para todas las tiendas
productos_tiendas = pd.concat([
    analizar_productos(tienda, 'Tienda 1'),
    analizar_productos(tienda2, 'Tienda 2'),
    analizar_productos(tienda3, 'Tienda 3'),
    analizar_productos(tienda4, 'Tienda 4')
])

## 3. Visualizaci√≥n de productos m√°s vendidos
plt.figure(figsize=(14, 10))
g = sns.FacetGrid(productos_tiendas[productos_tiendas['Tipo'] == 'M√°s vendidos'],
                 col='Tienda', col_wrap=2, height=5, aspect=1.5)
g.map_dataframe(sns.barplot, x='Ventas', y='Producto', palette='Blues_d')
g.set_titles("{col_name}")
g.set_axis_labels("Unidades Vendidas", "")
g.fig.suptitle('Top 5 Productos M√°s Vendidos por Tienda', y=1.05, fontweight='bold')

# A√±adir etiquetas de valor
for ax in g.axes.flat:
    ax.xaxis.set_major_formatter(thousands_formatter)
    for p in ax.patches:
        ax.annotate(format_thousands(p.get_width(), None),
                   (p.get_width(), p.get_y() + p.get_height()/2),
                   ha='left', va='center', xytext=(5, 0), textcoords='offset points')

plt.tight_layout()
plt.show()

## 4. Visualizaci√≥n de productos menos vendidos
plt.figure(figsize=(14, 10))
g = sns.FacetGrid(productos_tiendas[productos_tiendas['Tipo'] == 'Menos vendidos'],
                 col='Tienda', col_wrap=2, height=5, aspect=1.5)
g.map_dataframe(sns.barplot, x='Ventas', y='Producto', palette='Reds_d')
g.set_titles("{col_name}")
g.set_axis_labels("Unidades Vendidas", "")
g.fig.suptitle('Top 5 Productos Menos Vendidos por Tienda', y=1.05, fontweight='bold')

# A√±adir etiquetas de valor
for ax in g.axes.flat:
    ax.xaxis.set_major_formatter(thousands_formatter)
    for p in ax.patches:
        ax.annotate(format_thousands(p.get_width(), None),
                   (p.get_width(), p.get_y() + p.get_height()/2),
                   ha='left', va='center', xytext=(5, 0), textcoords='offset points')

plt.tight_layout()
plt.show()

## 5. Tabla resumen comparativa
print("\nüìä Tabla Comparativa de Productos Destacados")

# Crear resumen comparativo
resumen_productos = productos_tiendas.pivot_table(
    index=['Tienda', 'Tipo'],
    columns='Producto',
    values='Ventas',
    aggfunc='sum'
).fillna(0)

display(resumen_productos.style
       .background_gradient(cmap='Blues', axis=1)
       .format('{:,.0f}')
       .set_caption('Resumen de Ventas por Producto y Tienda'))

## 6. An√°lisis de categor√≠as de productos menos vendidos
print("\nüîç An√°lisis de Categor√≠as de Productos Menos Vendidos")
productos_menos_vendidos = productos_tiendas[productos_tiendas['Tipo'] == 'Menos vendidos']

# Obtener categor√≠as para estos productos
def obtener_categorias(df, productos):
    return df[df['Producto'].isin(productos)][['Producto', 'Categor√≠a del Producto']].drop_duplicates()

categorias_menos_vendidas = pd.concat([
    obtener_categorias(tienda, productos_menos_vendidos[productos_menos_vendidos['Tienda'] == 'Tienda 1']['Producto']),
    obtener_categorias(tienda2, productos_menos_vendidos[productos_menos_vendidos['Tienda'] == 'Tienda 2']['Producto']),
    obtener_categorias(tienda3, productos_menos_vendidos[productos_menos_vendidos['Tienda'] == 'Tienda 3']['Producto']),
    obtener_categorias(tienda4, productos_menos_vendidos[productos_menos_vendidos['Tienda'] == 'Tienda 4']['Producto'])
])
def obtener_categorias_con_tienda(df, productos, nombre_tienda):
    return (
        df[df['Producto'].isin(productos)]
        [['Producto', 'Categor√≠a del Producto']]
        .drop_duplicates()
        .assign(Tienda=nombre_tienda)  # A√±adimos la columna Tienda
    )

# Obtenemos categor√≠as con informaci√≥n de tienda
categorias_menos_vendidas = pd.concat([
    obtener_categorias_con_tienda(tienda,
                                productos_menos_vendidos[productos_menos_vendidos['Tienda'] == 'Tienda 1']['Producto'],
                                'Tienda 1'),
    obtener_categorias_con_tienda(tienda2,
                                productos_menos_vendidos[productos_menos_vendidos['Tienda'] == 'Tienda 2']['Producto'],
                                'Tienda 2'),
    obtener_categorias_con_tienda(tienda3,
                                productos_menos_vendidos[productos_menos_vendidos['Tienda'] == 'Tienda 3']['Producto'],
                                'Tienda 3'),
    obtener_categorias_con_tienda(tienda4,
                                productos_menos_vendidos[productos_menos_vendidos['Tienda'] == 'Tienda 4']['Producto'],
                                'Tienda 4')
])

# Merge correcto (ahora ambas tablas tienen 'Producto' y 'Tienda')
resultado_final = categorias_menos_vendidas.merge(
    productos_menos_vendidos[['Producto', 'Tienda', 'Ventas', 'Tipo']],
    on=['Producto', 'Tienda']
).sort_values(['Tienda', 'Ventas'])

# Mostramos resultados
display(resultado_final.style.set_caption('Productos menos vendidos con sus categor√≠as'))
```

# 5. Env√≠o promedio por tienda

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from matplotlib.ticker import FuncFormatter

# Configuraci√≥n de estilos
sns.set_theme(style="whitegrid")
plt.rcParams['figure.facecolor'] = 'white'

## 1. C√°lculo de m√©tricas de env√≠o por tienda
def calcular_costos_envio():
    # Diccionario para almacenar resultados
    resultados = {
        'Tienda': ['Tienda 1', 'Tienda 2', 'Tienda 3', 'Tienda 4'],
        'Costo Promedio Env√≠o': [
            tienda['Costo de env√≠o'].mean(),
            tienda2['Costo de env√≠o'].mean(),
            tienda3['Costo de env√≠o'].mean(),
            tienda4['Costo de env√≠o'].mean()
        ],
        'Costo Total Env√≠o': [
            tienda['Costo de env√≠o'].sum(),
            tienda2['Costo de env√≠o'].sum(),
            tienda3['Costo de env√≠o'].sum(),
            tienda4['Costo de env√≠o'].sum()
        ],
        'Cantidad de Env√≠os': [
            len(tienda),
            len(tienda2),
            len(tienda3),
            len(tienda4)
        ]
    }

    return pd.DataFrame(resultados)

# Calculamos las m√©tricas
costos_envio_df = calcular_costos_envio()
costos_envio_df['Costo Promedio Env√≠o'] = costos_envio_df['Costo Promedio Env√≠o'].round(2)
costos_envio_df['Costo Total Env√≠o'] = costos_envio_df['Costo Total Env√≠o'].round(2)

## 2. Visualizaci√≥n de costos promedio
plt.figure(figsize=(12, 6))
ax = sns.barplot(
    data=costos_envio_df,
    x='Tienda',
    y='Costo Promedio Env√≠o',
    palette='coolwarm',
    errorbar=None
)

# A√±adir etiquetas de valor
for p in ax.patches:
    ax.annotate(
        f"${p.get_height():,.2f}",
        (p.get_x() + p.get_width() / 2., p.get_height()),
        ha='center', va='center',
        xytext=(0, 10),
        textcoords='offset points',
        fontweight='bold'
    )

plt.title('Costo Promedio de Env√≠o por Tienda', fontsize=16, fontweight='bold', pad=20)
plt.xlabel('Tienda', labelpad=10, fontweight='bold')
plt.ylabel('Costo Promedio (USD)', labelpad=10, fontweight='bold')
plt.xticks(fontsize=12)
plt.yticks(fontsize=12)
sns.despine()
plt.tight_layout()
plt.show()

## 3. An√°lisis comparativo
plt.figure(figsize=(14, 7))
# Gr√°fico combinado de barras (total) y l√≠nea (promedio)
ax1 = sns.barplot(
    data=costos_envio_df,
    x='Tienda',
    y='Costo Total Env√≠o',
    color='skyblue',
    label='Costo Total'
)

ax2 = ax1.twinx()
sns.lineplot(
    data=costos_envio_df,
    x='Tienda',
    y='Costo Promedio Env√≠o',
    color='crimson',
    marker='o',
    markersize=10,
    linewidth=3,
    ax=ax2,
    label='Costo Promedio'
)

# A√±adir etiquetas y leyenda
ax1.set_ylabel('Costo Total Env√≠o (USD)', fontweight='bold')
ax2.set_ylabel('Costo Promedio (USD)', fontweight='bold')
ax1.set_xlabel('Tienda', fontweight='bold')
plt.title('Comparaci√≥n Costos de Env√≠o: Total vs Promedio', fontsize=16, fontweight='bold', pad=20)

# Combinar leyendas
lines, labels = ax1.get_legend_handles_labels()
lines2, labels2 = ax2.get_legend_handles_labels()
ax2.legend(lines + lines2, labels + labels2, loc='upper right')

# Formatear ejes
ax1.yaxis.set_major_formatter(FuncFormatter(lambda x, _: f"${x:,.0f}"))
ax2.yaxis.set_major_formatter(FuncFormatter(lambda x, _: f"${x:,.2f}"))

plt.tight_layout()
plt.show()

## 4. Tabla resumen detallada
print("\nüìä Resumen Anal√≠tico de Costos de Env√≠o")
display(costos_envio_df.style
       .format({
           'Costo Promedio Env√≠o': '${:,.2f}',
           'Costo Total Env√≠o': '${:,.2f}',
           'Cantidad de Env√≠os': '{:,}'
       })
       .background_gradient(subset=['Costo Promedio Env√≠o'], cmap='YlOrRd')
       .background_gradient(subset=['Costo Total Env√≠o'], cmap='Blues')
       .set_caption('M√©tricas de Costos de Env√≠o por Tienda'))

## 5. An√°lisis de distribuci√≥n (boxplot)
plt.figure(figsize=(12, 6))
sns.boxplot(
    data=pd.concat([
        tienda.assign(Tienda='Tienda 1'),
        tienda2.assign(Tienda='Tienda 2'),
        tienda3.assign(Tienda='Tienda 3'),
        tienda4.assign(Tienda='Tienda 4')
    ]),
    x='Tienda',
    y='Costo de env√≠o',
    palette='Set2'
)
plt.title('Distribuci√≥n de Costos de Env√≠o por Tienda', fontsize=16, fontweight='bold', pad=20)
plt.xlabel('Tienda', labelpad=10, fontweight='bold')
plt.ylabel('Costo de env√≠o (USD)', labelpad=10, fontweight='bold')
plt.gca().yaxis.set_major_formatter(FuncFormatter(lambda x, _: f"${x:,.0f}"))
sns.despine()
plt.tight_layout()
plt.show()
```

```python
# Creaci√≥n de un dashboard con subplots
fig = plt.figure(figsize=(18, 12))
gs = fig.add_gridspec(2, 2)

# Gr√°fico 1: Ingresos por tienda
ax1 = fig.add_subplot(gs[0, 0])
ingresos_tiendas = [tienda['Precio'].sum(), tienda2['Precio'].sum(),
                    tienda3['Precio'].sum(), tienda4['Precio'].sum()]
sns.barplot(x=['Tienda 1', 'Tienda 2', 'Tienda 3', 'Tienda 4'],
            y=ingresos_tiendas, palette='rocket', ax=ax1)
ax1.set_title('Ingresos Totales por Tienda', fontweight='bold')
ax1.set_ylabel('Ingresos (USD)')
ax1.yaxis.set_major_formatter(FuncFormatter(lambda x, _: f"${x/1e6:,.1f}M"))

# Gr√°fico 2: Costos de env√≠o promedio
ax2 = fig.add_subplot(gs[0, 1])
costos_envio = [tienda['Costo de env√≠o'].mean(), tienda2['Costo de env√≠o'].mean(),
                tienda3['Costo de env√≠o'].mean(), tienda4['Costo de env√≠o'].mean()]
sns.barplot(x=['Tienda 1', 'Tienda 2', 'Tienda 3', 'Tienda 4'],
            y=costos_envio, palette='mako', ax=ax2)
ax2.set_title('Costo Promedio de Env√≠o por Tienda', fontweight='bold')
ax2.set_ylabel('Costo (USD)')
ax2.yaxis.set_major_formatter(FuncFormatter(lambda x, _: f"${x:,.0f}"))

# Gr√°fico 3: Calificaciones promedio
ax3 = fig.add_subplot(gs[1, 0])
calificaciones = [tienda['Calificaci√≥n'].mean(), tienda2['Calificaci√≥n'].mean(),
                  tienda3['Calificaci√≥n'].mean(), tienda4['Calificaci√≥n'].mean()]
sns.barplot(x=['Tienda 1', 'Tienda 2', 'Tienda 3', 'Tienda 4'],
            y=calificaciones, palette='flare', ax=ax3)
ax3.set_title('Calificaci√≥n Promedio por Tienda', fontweight='bold')
ax3.set_ylabel('Puntuaci√≥n (1-5)')
ax3.set_ylim(0, 5)

# Gr√°fico 4: M√©todo de pago m√°s popular
ax4 = fig.add_subplot(gs[1, 1])
metodo_popular = pd.concat([tienda, tienda2, tienda3, tienda4])['M√©todo de pago'].value_counts()
sns.barplot(x=metodo_popular.index, y=metodo_popular.values, palette='crest', ax=ax4)
ax4.set_title('M√©todos de Pago M√°s Utilizados', fontweight='bold')
ax4.set_ylabel('Cantidad de Transacciones')
ax4.tick_params(axis='x', rotation=45)

plt.tight_layout()
plt.show()
```
